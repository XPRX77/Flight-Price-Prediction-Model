# Exploratory Data Analysis

In the above you would have seen that some columns are in wrong datatype so now we will put those in correct form
so that the model can understand and can be used for prediction


since the Date_of_Journey is object datatype so we will convert it into timestamps

```python
train_df["Journey_day"] = pd.to_datetime(train_df.Date_of_Journey,format = "%d/%m/%Y").dt.day
train_df['Journey_month'] = pd.to_datetime(train_df.Date_of_Journey,format ="%d/%m/%Y").dt.month
```

```python
train_df.head()
```

```python
#we will remove the Date_of_journey column as we took all usefull data from it
train_df.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
#similarly we wil take out the dep hour and dep min and drop the dep_time column
train_df["Dep_Hr"] = pd.to_datetime(train_df.Dep_Time).dt.hour
train_df["Dep_Min"] = pd.to_datetime(train_df.Dep_Time).dt.minute
train_df.drop(["Dep_Time"],axis = 1, inplace = True)
```

```python
train_df.head()
```

```python
#similarly we wil take out the Arrival min and arrival hr and then remove the arrival time column
train_df["Arrival_Hr"] = pd.to_datetime(train_df.Arrival_Time).dt.hour
train_df["Arrival_Min"] = pd.to_datetime(train_df.Arrival_Time).dt.minute
train_df.drop(["Arrival_Time"],axis = 1, inplace = True)
```

```python
train_df.head()
```

```python
# Now we will take out the duration hr and duration min from the Duration_Time
duration = list(train_df["Duration"])

#since few of the time were in '12h' or '55m' but all should be in the form of '10h 3m' i.e. hour and minute form
for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   #adds "0m" in the end 
        else:
            duration[i] = "0h " + duration[i]           #adds "0h" in the begining

duration_hr = []
duration_min = []
for i in range(len(duration)):
    duration_hr.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_min.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
    
```

```python
#Creating duration hr and duration min column
train_df['Duration_hr'] = duration_hr
train_df['Duration_min'] = duration_min
```

```python
# now we will drop the duration column 
train_df.drop(['Duration'], axis = 1, inplace = True)
```

# Categorical data handling

Nominal variable - Without any particular implied order like say color of eyes- Black, blue, green, brown. There is no ranking or ordering to the colors we use <b>*OneHotEncoder* </b>

Ordinal variable - Are categories with a natural ordering but the scale of difference is not defined. Example - Speed of a moving object - slow, medium, fast, super fast and ultra - fast. There is a natural ordering in it Slow < Medium < Fast < Super Fast < Ultra fast) We use <b>*labelEncoder*</b>

```python
#as airline is nominal data we will use onehotencoding
Airline = train_df[["Airline"]]

Airline = pd.get_dummies(Airline, drop_first= True)

Airline.head()
```

```python
train_df.Source.value_counts()
```

```python
#now we will check the price for each source

sns.catplot(x='Source', y='Price', data = train_df.sort_values("Price",ascending =False),kind = 'boxen')
```

```python
#source is a nominal data we will use OneHotEncoding
Source = train_df[['Source']]
Source = pd.get_dummies(Source, drop_first=True)
Source.head()
```

```python
#Destination is a nominal data we will use OneHotEncoding
Destination = train_df[["Destination"]]
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
train_df['Additional_Info'].value_counts()
```

```python
#from the above we can see most of the values in Additional_Info is "NO Info" so we will drop it
# Route and Total Stops are related to each other so we will drop the route column

train_df.drop(['Route','Additional_Info'], axis = 1, inplace = True)
```

```python
train_df.head()
```

```python
train_df['Total_Stops'].value_counts()
```

```python
#Since this is an ordinal data we will convert this using LabelEncoder
train_df.replace({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops': 3, '4 stops': 4}, inplace = True)
```
